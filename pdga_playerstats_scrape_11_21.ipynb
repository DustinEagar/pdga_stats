```python
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
import os
from tqdm.notebook import trange, tqdm
```

```python

test_url = "https://www.pdga.com/players/stats?Year=2021&player_Class=All&Gender=All&Bracket=All&continent=All&Country=All&StateProv=All"

#pretend to be a browser
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(test_url, headers=header)

test_tables = pd.read_html(r.text)

test_tables[0]
```

```python
'''
first page
https://www.pdga.com/players/stats?Year=2021&player_Class=All&Gender=All&Bracket=All&continent=All&Country=All&StateProv=All

second page
https://www.pdga.com/players/stats?Year=2021&player_Class=All&Gender=All&Bracket=All&continent=All&Country=All&StateProv=All&page=1

last page
https://www.pdga.com/players/stats?Year=2021&player_Class=All&Gender=All&Bracket=All&continent=All&Country=All&StateProv=All&page=6626

page=0 works for first page
iterate url pages over range(6627)
november: 6715 pages
'''
now_string = datetime.now().strftime("%d_%m_%y_%H-%M-%S")
filename = "pdga_player_stats_{}.csv".format(now_string)


df = pd.DataFrame()

for i in trange(6715):
    
    while True:
        try:
            time.sleep(3*np.random.beta(2,1,1)[0]) #Not strictly necessary, but good practice and polite
            url = "https://www.pdga.com/players/stats?Year=2021&player_Class=All&Gender=All&Bracket=All&continent=All&Country=All&StateProv=All&page={}".format(i)

            r = requests.get(url, headers=header)
            df1 = pd.read_html(r.text)[0] #take 0th item of tables list

            df = pd.concat([df, df1], ignore_index=True)
                
        except Exception as e:
            print("page {} failed with exception {}".format(i, e))
            time.sleep(30)
            continue
        break

df.to_csv(filename)
display(df)
    
```
